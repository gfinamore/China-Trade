```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
china2000 = pd.read_csv('china_2000.csv')
china2010 = pd.read_csv('china_2010.csv')
china2020 = pd.read_csv('china_2020.csv')

#print(china2000.dtypes)
#print(china2010.dtypes)
#print(china2020.dtypes)
```

```python
china_trade = pd.concat([china2000, china2010, china2020], ignore_index=True)
display(china_trade)
```

## Plot exploratório

- TS da balança comercial, (exportação e importação da china)

- Evolução do volume para cada país (exportação e importação)

- Evolução do volume para cada produto

- Evolução maiores déficits e supervits



```python
# primeiro gráfico

total_volume = china_trade.groupby('year')[['export_value','import_value']].sum().reset_index()

superavit = total_volume['export_value'] - total_volume['import_value']
total_volume['superavit'] = superavit

fig = px.line(
    total_volume,
    x='year',
    y=['export_value', 'import_value', 'superavit'],
    title='Exportações e Importações da China (2000-2024)',
    labels={
        'year': '',
        'value': '',
        'variable': 'Fluxo Comercial'
    },
    markers=True
)

fig.data[0].name = "Exportação"
fig.data[1].name = "Importação"
fig.data[2].name = "Superávit"

fig.show()
```

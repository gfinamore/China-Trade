```python
#relação com eua em gráfico

import pandas as pd
import plotly.graph_objects as go

df = pd.read_csv("balanca_g20_china_2000_2024.csv")

df_usa = df[df["country_iso3_code"] == "USA"].sort_values("year")


fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_usa["year"],
    y=df_usa["export_value"] / 1e9,
    mode='lines+markers',
    name="China's imports",
    line=dict(color="#1f77b4", width=2.5),
    marker=dict(symbol='circle', size=8)))

fig.add_trace(go.Scatter(
    x=df_usa["year"],
    y=df_usa["import_value"] / 1e9,
    mode='lines+markers',
    name="China's exports",
    line=dict(color="#d62728", width=2.5),
    marker=dict(symbol='square', size=8)))


fig.update_layout(
    title=dict(
        text="<b>Evolution of trade between China and USA (2000 - 2024)</b>",
        font=dict(size=16, color="#333333"),
        x=0.0, 
        xanchor='left'
    ),
    xaxis=dict(
        title="Year",
        title_font=dict(size=12, color="#333333"),
        tickmode='linear',
        tick0=df_usa["year"].min() if not df_usa.empty else 2000,
        dtick=1,
        tickangle=-45, 
        tickfont=dict(size=10, color="#333333"),
        showgrid=True,
        gridcolor='#e5e5e5'  
    ),
    yaxis=dict(
        title="USD Billions",
        title_font=dict(size=12, color="#333333"),
        showgrid=True,
        gridcolor='#e5e5e5'
    ),
    legend=dict(
        x=0.01,
        y=0.99,
        xanchor='left',
        yanchor='top',
        font=dict(size=11),
        bgcolor='rgba(255, 255, 255, 0.7)', 
        bordercolor='rgba(0, 0, 0, 0)'
    ),
    template="plotly_white",  
    width=1000,
    height=500,
    margin=dict(l=60, r=40, t=80, b=60)
)


fig.show()
```

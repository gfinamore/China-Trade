```python
#Relação com EUA em gráfico


import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("balanca_g20_china_2000_2024.csv")

df_usa = df[df["country_iso3_code"] == "USA"].sort_values("year")



#Grafico
sns.set_theme(style="whitegrid")


fig, ax = plt.subplots(figsize=(12, 6))


ax.plot(
    df_usa["year"],
    df_usa["export_value"] / 1e9,
    marker="o",
    linewidth=2.5,
    color="#1f77b4",
    label="Exportações dos EUA para a China (X)")

ax.plot(
    df_usa["year"],
    df_usa["import_value"] / 1e9,
    marker="s",
    linewidth=2.5,
    color="#d62728",
    label="Importações dos EUA da China (M)")


ax.set_title(
    "Evolução do Comércio entre EUA e China (2000 - 2024)",
    fontsize=16,
    fontweight="bold",
    pad=15)


ax.set_xlabel("Ano", fontsize=12)
ax.set_ylabel("Valor Comercial (Em Bilhões de USD)", fontsize=12)



#configurações de espaçamento do gráfico
ax.set_xticks(df_usa["year"])
ax.set_xticklabels(df_usa["year"], rotation=45)

ax.legend(fontsize=11, loc="upper left")

plt.tight_layout()

plt.savefig("evolucao_comercio_eua_china.png", dpi=300)
```

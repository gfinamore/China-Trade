```python
import pandas as pd
import plotly.express as px


df = pd.read_csv('hs92_gics.csv')
df_produtos = df[df['level'] == '4digit']

#quantidade por setor
contagem_setores = df_produtos['gics_sector'].value_counts().reset_index()
contagem_setores.columns = ['Sector - GICS', 'Quantity of Products']

#ordenação de quantidade
contagem_setores = contagem_setores.sort_values(by='Quantity of Products', ascending=True)

#grafico
fig = px.bar(
    contagem_setores, 
    x='Quantity of Products', 
    y='Sector - GICS',  # <-- Ajustado para o nome correto em inglês
    orientation='h',
    title='Quantity of Products by Sector (GICS)', # <-- Corrigido o erro de digitação
    text='Quantity of Products',
    color='Quantity of Products',
    color_continuous_scale='Blues')

fig.update_layout(
    yaxis_title='Sector - GICS',
    xaxis_title='Total of Products',
    showlegend=False,
    height=500)

fig.show()
```

```python
import pandas as pd

df = pd.read_csv("hs92_country_country_year.csv")

df.info()

df.head
```

```python
#filtrando período, países do g20 e china

g20_countries = [
    "ARG",
    "AUS",
    "BRA",
    "CAN",
    "FRA",
    "DEU",
    "IND",
    "IDN",
    "ITA",
    "JPN",
    "KOR",
    "MEX",
    "RUS",
    "SAU",
    "ZAF",
    "TUR",
    "GBR",
    "USA",
]

filtro = (
    (df["year"] >= 2000)
    & (df["year"] <= 2024)
    & (df["partner_iso3_code"] == "CHN")
    & (df["country_iso3_code"].isin(g20_countries))
)


df_filtrado = df[filtro].copy()


#balança
df_filtrado["balanca_comercial"] = (df_filtrado["export_value"] - df_filtrado["import_value"])


#salvando
df_filtrado = df_filtrado.sort_values(by=["country_iso3_code", "year"])

df_filtrado.to_csv("balanca_g20_china_2000_2024.csv", index=False)
df_filtrado.to_excel("balanca_g20_china_2000_2024.xlsx", index=False)

print("Análise da Balança Comercial dos países do G20 com a China, de 2000 a 2024.")
print(df_filtrado.head())
```

```python
#países com maior relação

df_filtrado["comercio"] = (
    df_filtrado["export_value"] + df_filtrado["import_value"]
)


ranking_g20 = (
    df_filtrado.groupby("country_iso3_code")["comercio"]
    .sum()
    .sort_values(ascending=False)
)

print("Ranking de relação com a China de 2000 a 2024")
print(ranking_g20.head())
print(ranking_g20.tail())
```
